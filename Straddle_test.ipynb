```python
def day(date):
    date = datetime.strptime(date,'%d-%m-%Y')
    return date.strftime('%A')
```

```python
def get_atm(spot):
    if spot%100<50:
        return int(spot/100)*100
    return int(spot/100)*100 + 100
```
